# Orientation Test

```python
import numpy as np
import matplotlib.pyplot as plt

def orientation_test(p1, p2, p3):
    """Test the orientation of three points."""
    u = p2 - p1
    v = p3 - p1

    return np.cross(u, v)
```

```python
points = np.random.rand(6, 2)

# Sort by x-value
points = points[np.argsort(points[:, 0])]

# Plot the points
plt.plot(points[:, 0], points[:, 1], 'o')

# Create line segments between the left 3 points
for i in range(2):
    plt.plot([points[i, 0], points[i+1, 0]], [points[i, 1], points[i+1, 1]], 'k-')
plt.plot([points[2, 0], points[0, 0]], [points[2, 1], points[0, 1]], 'k-')

# Create line segments between the right 3 points
for i in range(3, 5):
    plt.plot([points[i, 0], points[i+1, 0]], [points[i, 1], points[i+1, 1]], 'k-')
plt.plot([points[5, 0], points[3, 0]], [points[5, 1], points[3, 1]], 'k-')
```

```python
S1 = points[:3]
S2 = points[3:]

# Select the right-most point of S1 and the left-most point of S2
p1 = S1[np.argmax(S1[:, 0])]
p2 = S2[np.argmin(S2[:, 0])]

test_point = points[4]

# Test the orientation of the points
orientation = orientation_test(p1, p2, test_point)
print(orientation)
```
